```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import seaborn as sn
```

```python
sn.set_theme(style="darkgrid")
x = np.linspace(-2*np.pi, 2*np.pi, 2500)
y = -20.0*np.exp(-0.2*np.sqrt(0.5*(x**2)))-np.exp(0.5*(np.cos(2*np.pi*x)))+20

plt.plot(x, y)
plt.title('Optimization Landscape')
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

```python
def f(x):
  """y=f(x)"""
  return -20.0*np.exp(-0.2*np.sqrt(0.5*(x**2)))-np.exp(0.5*(np.cos(2*np.pi*x)))+20

def dfdx(x, dx=1e-3):
  """y'=dy(x)/dx"""
  return (f(x+dx)-f(x))/dx

def GD_update(x, lr=4e-3):
  """vanilla gradient descent update;
     x(new) = x(old) - lr*dy(x)/dx"""
  gradient = dfdx(x)
  return (x-lr*gradient)

def MomentumGD_update(xo, vo, beta=0.99, lr=4e-2):
  """gradient descent with momentum update;
     v(new) = b*v(old) + (1-b)*dy(x)/dx
     x(new) = x(old) - lr*dy(x)/dx"""
  vn = beta*vo + (1-beta)*dfdx(xo)
  xn = xo - lr * vn
  return xn, vn

def visualizer(xc):
  yc = f(xc)
  plt.plot(x, y)
  plt.plot(xc, yc, 'o')
  plt.title('Optimization Landscape')
  plt.xlabel("Independent Variable")
  plt.ylabel("Dependent Variable")
  plt.show()

def simulate(x0, num_iterations, update_rule):
  history = np.zeros((num_iterations, 2))
  v0=0
  if update_rule == "GD":
    for i in range(num_iterations):
      history[i] = [x0, f(x0)]
      x0 = GD_update(x0)

  if update_rule == "MGD":
    for i in range(num_iterations):
      history[i] = [x0, f(x0)]
      x0, v0 = MomentumGD_update(x0, v0)

  return history
```

```python
x0=5.5
num_iterations=1500
update_rule='GD'
gd_history = simulate(x0, num_iterations, update_rule)

update_rule='MGD'
mgd_history = simulate(x0, num_iterations, update_rule)
```
